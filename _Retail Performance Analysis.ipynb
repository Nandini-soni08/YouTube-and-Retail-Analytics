```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
retail_df = pd.read_csv('Superstore.csv', encoding='latin1')
retail_df
```

```python
# Clean Data
```

```python
retail_df.dropna(inplace=True)
```

```python
#  Profit by Category
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Profit', data=retail_df, estimator=sum)
plt.title("Total Profit by Category")
plt.xticks(rotation=45)
plt.show()
```

```python
# Sales by Region
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Region', y='Sales', data=retail_df, estimator=sum)
plt.title("Total Sales by Region")
plt.show()
```

```python
# Monthly Sales Trend
```

```python
retail_df['Order Date'] = pd.to_datetime(retail_df['Order Date'])
retail_df['Month'] = retail_df['Order Date'].dt.to_period('M')
monthly_sales = retail_df.groupby('Month')['Sales'].sum()
```

```python
monthly_sales.plot(kind='line', figsize=(10,5), marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
